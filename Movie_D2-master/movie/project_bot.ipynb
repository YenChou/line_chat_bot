```python
"""

啟用伺服器基本樣板

"""

# 引用Web Server套件
from flask import Flask, request, abort

# 從linebot 套件包裡引用 LineBotApi 與 WebhookHandler 類別
from linebot import (
    LineBotApi, WebhookHandler
)

# 引用無效簽章錯誤
from linebot.exceptions import (
    InvalidSignatureError
)

# 載入json處理套件
import json

# 載入基礎設定檔
secretFileContentJson=json.load(open("line_secret_key",'r'))
server_url=secretFileContentJson.get("server_url")

# 設定Server啟用細節
app = Flask(__name__,static_url_path = "/images" , static_folder = "../images/")

# 生成實體物件
line_bot_api = LineBotApi(secretFileContentJson.get("channel_access_token"))
handler = WebhookHandler(secretFileContentJson.get("secret_key"))

# 啟動server對外接口，使Line能丟消息進來
@app.route("/", methods=['POST'])
def callback():
    # get X-Line-Signature header value
    signature = request.headers['X-Line-Signature']

    # get request body as text
    body = request.get_data(as_text=True)
    app.logger.info("Request body: " + body)

    # handle webhook body
    try:
        handler.handle(body, signature)
    except InvalidSignatureError:
        abort(400)

    return 'OK'
```

```python
import mysql.connector
import time
config_sql = {
        'host': '10.120.28.11',
        'port': 3306,
        'database': 'project_product',
        'user': 'root',
        'password': 'root',
        'charset': 'utf8'
    }

def update_inventory_transaction(config, data):
    inventory_colname = 'num'
    table_name = 'product'
    db=None
    cursor=None
    if data:
        try:
            db = mysql.connector.Connect(**config,auth_plugin= 'mysql_native_password')
            cursor = db.cursor()
            db.autocommit = False
            db.start_transaction()
            #UPDATE "表格名" SET "欄位1" = [新值] WHERE "條件" ;
            for pid in set(data):
                if data.count(pid) > 1:
                    stmt_update_many = 'UPDATE {} SET {} = {} - {} WHERE name = "{}" ;'.format(table_name,
                                                                                              inventory_colname,
                                                                                              inventory_colname,
                                                                                              data.count(pid),
                                                                                              pid)
                    cursor.execute(stmt_update_many)
                    # set(data).remove(pid)
                else:
                    stmt_update_one = 'UPDATE {} SET {} = {} - 1 WHERE name = "{}" ;'.format(table_name,
                                                                                            inventory_colname,
                                                                                            inventory_colname,
                                                                                            pid)
                    cursor.execute(stmt_update_one)
                    # time.sleep(10)
            db.commit()
            return True
    
        except mysql.connector.Error as a:
            db.rollback()
            return a
        finally:
            db.close()
            cursor.close()
    else:
        return False
```

```python
#用戶點擊button後，觸發postback event，對其回傳做相對應處理
@handler.add(PostbackEvent)
def handle_post_message(event):
    user_profile = line_bot_api.get_profile(event.source.user_id)
    if (event.postback.data.find('result_ready')== 0):   
        line_bot_api.reply_message(
        event.reply_token,
        buttons_template_counter)
    elif (event.postback.data.find('counter_1')== 0):
        #get face info
        kafka = 1
        if kafka:
            line_bot_api.reply_message(
            event.reply_token,
            buttons_template_product)
        else:
            line_bot_api.reply_message(
            event.reply_token,
            TextSendMessage(text="No detected,click the 'Paid program' button to restart"))
    elif (event.postback.data.find('counter_2')== 0):
            line_bot_api.reply_message(
            event.reply_token,
            TextSendMessage(text="只有櫃台 1 ~"))   
    elif (event.postback.data.find('counter_3')== 0):
            line_bot_api.reply_message(
            event.reply_token,
            TextSendMessage(text="只有櫃台 1 ~"))   
    elif (event.postback.data.find('counter_4')== 0):
            line_bot_api.reply_message(
            event.reply_token,
            TextSendMessage(text="只有櫃台 1 ~"))   

    elif (event.postback.data.find('result_product_yes') == 0):
            #get the product info  ***
            redis_result = redis_get_product()
            if redis_result:
                text = ",".join(product_list)
                line_bot_api.reply_message(
                event.reply_token,
                [buttons_template_paid,TextSendMessage(text=text)])
            else:
                line_bot_api.reply_message(
                event.reply_token,
                TextSendMessage(text="redis get data failed"))
    elif (event.postback.data.find('result_product_no') == 0):
            line_bot_api.reply_message(
            event.reply_token,
            TextSendMessage(text="End, click the 'Paid program' button to restart"))
            
# Agree transaction
    elif (event.postback.data.find('result_yes') == 0):
            #write to Mysql ***
            #produce to kafka ***
            
            status = update_inventory_transaction(config_sql, product_list)
            if status:
                topic = "transaction"
                bootstrap_servers = "master:6667"
                producer = KafkaProducer(bootstrap_servers=bootstrap_servers)
                result = {"memberID":str(event.source.user_id),
                         'food': product_list}
                result_final=json.dumps(result).encode('utf-8')
                producer.send(topic,value=result_final)
                line_bot_api.reply_message(
                event.reply_token,
                TextSendMessage(text="Complete !!! Thanks for comming, see you !!!"))
                delete_menu(event.source.user_id)
            else:
                line_bot_api.reply_message(
                event.reply_token,
                TextSendMessage(text="Sql failed, please restart !!!"))
            
            
    elif (event.postback.data.find('result_repeat') == 0):
            repeat_list = [TextSendMessage(text="Re-scaning..."), buttons_template_product]
            line_bot_api.reply_message(
            event.reply_token,
            repeat_list)
    elif (event.postback.data.find('result_no') == 0):
            line_bot_api.reply_message(
            event.reply_token,
            TextSendMessage(text="End, click the 'Paid program' button to restart"))
    else:
        pass
```
